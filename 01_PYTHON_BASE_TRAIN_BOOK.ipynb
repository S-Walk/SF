```python
""" -------------------------- Поменять местами крайние элементы списка """

def swap_places(lst):
    res = []
    res.append(lst[-1])
    res.extend(lst[1:-1])
    res.append(lst[0])
    
    return res

print(swap_places([1, 2, 3]))
## [3, 2, 1]
print(swap_places([1, 2, 3, 4, 5]))
## [5, 2, 3, 4, 1]
print(swap_places(['н', 'л', 'о', 'с']))
## ['с', 'л', 'о', 'н']
```

```python
""" ------------------------------------ Выравнивание блины строк """

def equalize_lengths(lst):
    lst.sort(key=len)
    max_string = len(lst[-1])
    lst = lst[::-1]
    res = []
    for string in lst:
        while len(string) < max_string:
            string += '_'
        res.append(string)
    
    return res

print(equalize_lengths(['крот', 'белка', 'выхухоль']))
print(equalize_lengths(['a', 'aa', 'aaa', 'aaaa', 'aaaaa']))
print(equalize_lengths(['qweasdqweas', 'q', 'rteww', 'ewqqqqq']))
```

```python
""" ---------------------------- Минимальное число и сумма минимальных чисел """

def find_min_number(a,b,c):
    min_num = a
    if b < min_num: min_num = b
    if c < min_num: min_num = c
    
    return min_num

def sum_min_numbers(a,b,c):
    min_num = find_min_number(a,b,c)
    num_lst = [a,b,c]
    num_lst.remove(min_num)
    if num_lst[0] < num_lst[1]: min_sum = min_num + num_lst[0]
    else: min_sum = min_num + num_lst[1]
    
    return min_sum

print(sum_min_numbers(1, 2, 3))
print(sum_min_numbers(1, 2, -10))
```

```python
""" ----------------------------------------- Счастливый билетик """

def lucky_ticket(number):
    half_1 = 0
    half_2 = 0
    for i in str(number)[:3]:
        half_1 +=int(i)
    for i in str(number)[3:]:
        half_2 +=int(i)
    if half_1 == half_2: return True
    else: return False

print(lucky_ticket(111111))
print(lucky_ticket(123456))
```

```python
""" ----------------------------------- Количество отверстий в числе """

dict_holes = {0: 1, 4: 1, 6: 1, 8: 2, 9: 1}

def holes_count(num):
    count = 0
    for i in str(num):
        if int(i) in dict_holes: count += dict_holes[int(i)]
    return count

print(holes_count(8))
print(holes_count(146))
print(holes_count(84628))
```

```python
""" ---------------------------------- Количество чётных чисел в матрице """

matrix_example = [
    [1, 5, 4],
    [4, 2, -2],
    [7, 65, 88]
]

def even_numbers_in_matrix(matrix):
    count = 0
    for i in matrix:
        for j in i:
            if j % 2 == 0: count +=1
    return count

print(even_numbers_in_matrix(matrix=matrix_example))
```

```python
""" ----------------------------------------- Сложение матриц"""

matrix1_example = [
          [1, 5, 4, 5],
          [4, 2, -2, -5],
          [4, 2, -2, -5]
]

matrix2_example = [
          [10, 15, 43],
          [41, 2, -2],
          [7, 5, 7]
]

def matrix_sum(matrix1, matrix2):
    if len(matrix1) != len(matrix2):
        print('Error! Matrices dimensions are different!')
        return
    for i in matrix1:
        for j in matrix2:
            if len(i) != len(j):
                print('Error! Matrices dimensions are different!')
                return
    
    res_matr = []
    for i in range(len(matrix1)):
        res_matr.append([])
        for j in range(len(matrix1[i])):
            res_matr[i].append(matrix1[i][j] + matrix2[i][j])
    return res_matr

print(matrix_sum(matrix1=matrix1_example, matrix2=matrix2_example))
```

```python
""" ------------------------------------ Использование MAP() """

def get_total_score(data):
    map_lambda = lambda tup: (*tup, tup[1]+tup[2]+tup[3])
    upd_data = list(map(map_lambda, data))
    upd_data.sort(key = lambda x: x[4])
    
    return upd_data

data = [
    ('Amanda', 37, 78, 67),
    ('Patricia', 78, 93, 68),
    ('Marcos', 79, 67, 89),
    ('Dmitry', 67, 68, 100),
    ('Andrey', 100, 78, 76),
    ('Victoria', 93, 69, 96),
]

print(get_total_score(data))
```

```python
""" -------------------------------------- Пересечение списков """

input_string1 = input('Введите 1-ую последовательность идентификаторов: ')
input_string2 = input('Введите 2-ую последовательность идентификаторов: ')

try:    # преобразуем введённое в списки чисел, одновременно проверяя корректность ввода
    list_1 = list(map(lambda x: int(x), input_string1.split(sep=', ')))
    list_2 = list(map(lambda x: int(x), input_string2.split(sep=', ')))
except ValueError:    # отлавливаем ошибку ввода, если есть, печатаем информацию об ошибке
    print('Некорректный ввод')
else:    # удаляем дубликаты, находим пересечение, выводим результат
    res = set(list_1).intersection(set(list_2))
    print(list(res))
```

```python
""" ----------------------------- Минимум и максимум последовательности """

input_string = input('Введите последовательность чисел: ')

def find_min_max(seq):
    seq = seq.replace(',', '.')    # заменяем запятые на точки
    list_seq = seq.split(' ')    # преобразуем строку в список, разделяя по пробелу
    for elem in list_seq:    # убираем из списка слова
        if elem.isalpha(): list_seq.remove(elem)
    fine_seq = list(map(lambda x: float(x), list_seq))    # перобразуем элементы списка в числа

    return min(fine_seq), max(fine_seq)    # возвращаем минимальное и максимальное значение

#  выводим результат
print(" Minimum: ", find_min_max(input_string)[0], '\n', "Maximum: ", find_min_max(input_string)[1])
```

```python
""" ------------------------------------- Перобразование слов в числа """

input_string = input('Введите число словами: ')
number_word_dict = {
    "ты": 1000, "м": 1000000,
    "сто": 100, "двес": 200, "трис": 300, "четырес": 400, "пятьс": 500, "шестьс": 600, "семьс": 700, "восемьс": 800, "девятьс": 900,
    "одинн": 11, "двен": 12, "трин": 13, "четырн": 14, "пятн": 15, "шестн": 16, "семн": 17, "восемн": 18, "девятн": 19,
    "двад": 20, "трид": 30, "сор": 40, "пятьд": 50, "шестьд": 60, "семьд": 70, "восемьд": 80, "девяно": 90,
    "дес": 10, "н": 0, "о": 1, "дв": 2, "т": 3, "ч": 4, "п": 5, "ш": 6, "с": 7, "в": 8, "д": 9, }

def transform_string_to_integer(string):
    word_list = input_string.split(' ')    # перобразуем строку в список, разделяя по пробелам
    res_num = 0    # объявляем переменные - счётчики
    mil_num = 0
    thous_num = 0
    for word in word_list:
        if word.startswith("ты"):    # ищем слово-"маркер" обозначающее тысячи в числе
            thous_num = res_num * number_word_dict.get("ты")    # заносим количество тысяч в thous_num
            res_num = 0    # обнуляем общий счётчик
        elif word.startswith("м"):    # проводим ту же операцию с миллионами
            mil_num = res_num * number_word_dict.get("м")
            res_num = 0                
        else:    
            for key in number_word_dict:    # ищем ключ, с которого начинается слово
                if word.startswith(key):
                    res_num += number_word_dict.get(key)    # плюсуем значение ключа к общему счётчику
                    break

    res_num += thous_num + mil_num    # складываем все счэтчики
                
    return res_num
            

print(transform_string_to_integer(input_string))                    
```

```python
""" ----------------------- Разложение факториала числа на простые множители """

def get_simple_num(n):    # Списока простых чисел до N (!!! осторожно, очень ресурсоёмкая!!!) 
    lst = []
    need_num = [2,4,6,8]
    for i in range(2, n+1):
        if i > 10:
            if int(str(i)[-1]) in need_num: continue
        for j in lst:
            if j > i ** 1/2:
                lst.append(i)
                break
            if i % j == 0: break
        else: lst.append(i)
    return lst

def fact(n):    # Факториал N
    if n == 1 or n == 0: return 1
    res = n * fact(n-1)
    return res

def decompos_factorial(n):
    number = fact(n)
    div_list = []
    div = 2
    while div * div <= number:
        if number % div == 0:
            div_list.append(div)
            number //= div
        else: div += 1
    if n > 1: div_list.append(number)
    
    div_set = sorted(list(set(div_list)))
    res = ''
    for i in div_set:
        degree = div_list.count(i)
        if degree == 1: res += str(i) + ' * '
        else: res += str(i) + '^' + str(degree) + ' * '
    res = res[:-3]
    return res

print(decompos_factorial(22))
```

```python
""" ------------------------------------ Замена дубликатов в строке """

import string
alphabet = string.ascii_lowercase

input_s = 'eastorwesthomeisbest'

def replace_duplicates(s, details=False):    # переменная details отвечает за вывод подробной информации
    let_list = list(s)    # преобразуем строку в список
    step = 0    # номер итерации (если details=True)
    while sorted(let_list) != sorted(list(set(let_list))):    # будем запускать цикл по елементам списка пока есть неуникальные элементы
        for let in let_list:    # запускаем цикл по списку букв
            while let_list.count(let) > 1:    # пока есть дубликаты
                if let == 'z': next_let = 'a'    # буква для замены (случай замены z -> a обрабатываем отдельно)
                else: next_let = alphabet[alphabet.index(let) + 1]    
                let_list.remove(let)    # изменяем исходный список чтобы подать его на вход цикла (WHILE - если есть ещё дубликаты текущего елемента FOR - если нужно переходить к следующему элементу)
                let_list.remove(let)
                let_list.append(next_let)
                if details:    # если нужен подробный вывод
                    step += 1
                    print(f'Step {step}: ', ''.join(let_list))
    res = ''.join(let_list)    # формируем результат работы функции
    
    return res

print(replace_duplicates(input_s, True))
```
