```python
from collections import Counter
```

```python
""" Counter - разновидность словаря, имеющая свои методы для удобной работы с данными """

c = Counter()
c['red'] = 3    # можно передать ключ:значение
c['red'] += 2
print(c)

cars = ['red', 'blue', 'black', 'black', 'black', 'red', 'blue', 'red', 'white']
c = Counter(cars)    # можно передать итерируемый объект
print(c)

print(sum(c.values()))    # сумма всех значений
print(c['black'])    # вернуть значение как в dict
print(c.values())    # список значений как в  dict

cars_moscow = ['black', 'black', 'white', 'black', 'black', 'white', 'yellow', 'yellow', 'yellow']
cars_spb = ['red', 'black', 'black', 'white', 'white', 'yellow', 'yellow', 'red', 'white']

counter_moscow = Counter(cars_moscow)
counter_spb = Counter(cars_spb)

print(counter_moscow + counter_spb)    # сумма
print(counter_moscow - counter_spb)    # разность без отрицательных
counter_moscow.subtract(counter_spb)    # полная разность (изменяет исходник -> переопределяем)
print(counter_moscow)

counter_moscow = Counter(cars_moscow)
counter_spb = Counter(cars_spb)

print(*counter_moscow.elements())    # список всех элементов в порядке встречаемости
print(list(counter_moscow))    # список уникальных эдементов
print(dict(counter_moscow))    # словарь из Counter
print(counter_moscow.most_common())    # список из кортежей элементов в порядке убывания их встречаемости
print(counter_moscow.most_common(2))    # можно вернуть самый частый
counter_moscow.clear()    # очистка
```

```python
from collections import defaultdict
```

```python
""" В defaultdict можно задать структуру, которая по умолчанию будет храниться в значениях ключей """

students = [('Ivanov',1),('Smirnov',4),('Petrov',3),('Kuznetsova',1),
            ('Nikitina',2),('Markov',3),('Pavlov',2)]

groups = defaultdict(list)    # зададим список в кач-ве структуры по умолчанию

for student, group in students:
    groups[group].append(student)    # если ключа нет, он создаётся с структурой по умолчанию
    
print(groups)

print(groups[3])    # значение по ключу (если нет вернётся структура по умолчанию)
```

```python
from collections import deque
```

```python
""" DEQUE - двухконцевая очередь """

dq = deque([1, 2, 3, 4])
print(dq)

dq.append('000')
dq.appendleft('000')
print(dq)

dq.extend([3,2,1])
dq.extendleft([3,2,1]) # разворачивает добавляемый список !!!!!
print(dq)

dq.pop()
dq.popleft()
print(dq)

del dq[0]
print(dq, '\n'*2)


limited = deque([1,3,4,5,6,7], maxlen=3)    # очередь с ограниченной максимальной длиной
print(limited)    # при любых изменениях в deque остаются только последние maxlen элементов

limited.extend([1,2,3])
print(limited)
 
limited.append(8)
print(limited, '\n'*2)


days = deque(maxlen=7)

""" Средняя температура за последние 7 дней для каждого дня """ 
temps = [20.6, 19.4, 19.0, 19.0, 22.1,
        22.5, 22.8, 24.1, 25.6, 27.0,
        27.0, 25.6, 26.8, 27.3, 22.5,
        25.4, 24.4, 23.7, 23.6, 22.6,
        20.4, 17.9, 17.3, 17.3, 18.1,
        20.1, 22.2, 19.8, 21.3, 21.3,
        21.9]

for temp in temps:
    days.append(temp)
    if len(days) == days.maxlen:
        print(round(sum(days) / len(days), 2), end='; ') # Напечатаем пустую строку, чтобы завершить действие параметра end. Иначе следующая строка окажется напечатанной на предыдущей
print("'\n"*2)


dq = deque([1,2,3,4,5])
dq.reverse()
print(dq)

dq.rotate(2)    # переносит элементы из конца в начало
print(dq)

dq.rotate(-2)    # переносит элементы из начала в конец
print(dq)

print(dq.index(4))    # ValueError если индекс out of range
print(dq.count(4))

dq.clear()
print(dq)
```

```python
""" ----------------- !!!!!  Если присвоить срезу значение, исходный массив изменится  !!!!! """
def get_chess(a):
    arr = np.zeros((a,a))
    arr[1::2, ::2] = 1
    arr[::2, 1::2] = 1
    
    return arr

print(get_chess(4))


def shuffle_seed(array):
    random_num = np.random.randint(2**32)
    np.random.seed(random_num)
    user_arr = np.random.permutation(array)
    
    return user_arr, random_num

print(shuffle_seed([1, 2, 3, 4, 5]))


def min_max_dist(*vectors):
    dist_list = []
    for vec1 in vectors:
        for vec2 in vectors:
            distance = np.linalg.norm(vec1 - vec2)
            if distance == 0: continue
            dist_list.append(distance)  
            
    return min(dist_list), max(dist_list)  

vec1 = np.array([1,2,3])
vec2 = np.array([4,5,6])
vec3 = np.array([7, 8, 9])

print(min_max_dist(vec1, vec2, vec3))


def get_loto(num):
    arr = np.random.randint(1,101, size=(num,5,5))
    
    return arr

print(get_loto(3))


def unique_loto(num):
    sample = np.arange(0, 101)
    res = []
    for i in range(num):
        res.append(np.random.choice(sample, size=(5,5), replace=False))
    res = np.array(res)
    return res

print(unique_loto(3))
```
