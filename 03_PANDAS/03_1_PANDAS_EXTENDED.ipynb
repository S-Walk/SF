## Сортировка
### данных в DataFrame с помощью SORT_VALUES()

- by — имя или список имён столбцов, по значениям которых производится сортировка.
- axis — ось, по которой производится сортировка (0 — строки, 1 — столбцы). По умолчанию сортировка производится по строкам.
- ascending — сортировка по возрастанию (от меньшего к большему). По умолчанию параметр выставлен на True, для сортировки по убыванию (от большего к меньшему) необходимо выставить его на False.
- ignore_index — создаются ли новые индексы в таблице. По умолчанию выставлен на False и сохраняет индексы изначальной таблицы.
- inplace — производится ли замена исходной таблицы на отсортированную. По умолчанию параметр выставлен на False, то есть замены не производится. Чтобы переопределить исходную таблицу на отсортированную, необходимо выставить этот параметр на True.

```python
display(melb_df.sort_values(by='Price').head(10))    # прямая сортировка по столбцу
display(melb_df.sort_values(by='Date', ascending=False))    # обратная сортировка

melb_df.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']]    # каждая 10-я строка сортированной по 2 столбцам таблицы
```

## Группировка
### данных с помощью метода GROUPBY()

- by — имя или список имён столбцов, по которым производится группировка.
- axis — ось, по которой производится группировка (0 — строки, 1 — столбцы). По умолчанию группировка производится по строкам.
- as_index — добавляется ли дополнительный индекс к таблице. По умолчанию установлен на True.

```python
# сгруппируем данные по типу недвижимости и найдём среднее в каждом числовом признаке
melb_df.groupby(by='Type').mean(numeric_only=True)

# срднее в конкретном столбце - получим Series
melb_df.groupby('Type')['Price'].mean()

# найдём удалённость районов от центра и отсортируем
melb_df.groupby('Regionname')['Distance'].min().sort_values(ascending=False)

# найдём количество продаж а так же среднюю и максимальную цены (с помощью ф-ии .AGG()) в зависимости от месяца
melb_df.groupby('MonthSale')['Price'].agg(
    ['count', 'mean', 'max']
).sort_values(by='count', ascending=False)

# если в AGG() передать 'describe', получим инф-ю обо всех основных статистических параметрах
melb_df.groupby('MonthSale')['Price'].agg('describe')

# найдём число уникальных продавцов для каждого региона ( + список продавцов с помощью ф-ии SET)
melb_df.groupby('Regionname')['SellerG'].agg(['nunique', set])

# найдём среднюю цену в зависимости от кол-ва комнат
melb_df.groupby('Rooms')['Price'].mean()

#  вычислим стандартное отклонение по широте для регилнов
melb_df.groupby('Regionname')['Lattitude'].std().sort_values()

# найдём компанию с наименьшей выручкой за период
period = melb_df[(melb_df['Date'] >= pd.to_datetime('2017-05-01')) & (melb_df['Date'] >= pd.to_datetime('2017-09-01'))]
period.groupby('SellerG')['Price'].sum().sort_values()
```

### Метод   PIVOT_TABLE()   для построения сводных таблиц

- values — имя столбца, по которому необходимо получить сводные данные, применяя агрегирующую функцию;
- index — имя столбца, значения которого станут строками сводной таблицы;
- columns — имя столбца, значения которого станут столбцами сводной таблицы;
- aggfunc — имя или список имён агрегирующих функций (по умолчанию — подсчёт среднего, 'mean');
- fill_value — значение, которым необходимо заполнить пропуски (по умолчанию пропуски не заполняются).

```python
# построим ту же таблицу
melb_df.pivot_table('Price', index='Rooms', columns='Type', fill_value=0).round(0)

# найдём кол-во продаж по региону в зависимости от выходного дня
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count'
)

# зависимость средней и медианной площади от регилна и типа здания
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

### метод CONCAT()

- objs — список объектов DataFrame ([df1, df2,…]), которые должны быть сконкатенированы;
- axis — ось определяет направление конкатенации: 0 — конкатенация по строкам (по умолчанию), 1 — конкатенация по столбцам;
- join — либо inner (пересечение), либо outer (объединение); рассмотрим этот момент немного позже;
- ignore_index — по умолчанию установлено значение False, которое позволяет значениям индекса оставаться такими, какими они были в исходных данных. Если установлено значение True, параметр будет игнорировать исходные значения и повторно назначать значения индекса в последовательном порядке.

```python
# объединим таблицы с рейтингами по строкам
ratings = pd.concat([rat_1, rat_2], ignore_index=True)
display(ratings)
print('Строк в таблице RATINGS: ', ratings.shape[0])
print('Строк в таблице DATES: ', dates.shape[0])
display(rat_1.tail(1))
display(rat_2.head(1))    # одна строка задвоилась

"""  избавимся от дубликатов с помощью DROP_DUPLICATES()  """

ratings = ratings.drop_duplicates(ignore_index=True)
print('Строк в таблице RATINGS: ', ratings.shape[0])

# объединим рейтинги с датами по столбцам
ratings_dates = pd.concat([ratings, dates], axis=1)
display(ratings_dates.tail(7))
```

```python
""" Ф-ия для объединения файлов из папки в общий DataFrame """

import os
import pandas as pd

def concat_user_files(path):
    all_data = pd.DataFrame()
    file_list = os.listdir(path)
    file_list.sort()
    
    for file in file_list:
        full_path = path + '/' + file
        file_data = pd.read_csv(full_path)
        all_data = pd.concat([all_data, file_data])
    
    all_data = all_data.drop_duplicates(ignore_index=True)
    
    return all_data
```

### метод JOIN ()

- other — таблица, которую мы присоединяем. При объединении она является «правой», а исходная таблица, от имени которой вызывается метод, является «левой».
- how — параметр типа объединения. Он может принимать значения 'inner', 'left' (left outer), 'right' (right outer), и 'outer' (full outer). По умолчанию параметр установлен на 'left'.
- on — параметр, который определяет, по какому столбцу в «левой» таблице происходит объединение по индексам из «правой».
- lsuffix и rsuffix — дополнения (суффиксы) к названиям одноимённых столбцов в «левой» и «правой» таблицах.

```python
# объединим таблицы по индексам (без ключегого столбца) с параметром LEFT, присвоим данным из DATA суффикс '_right'
joined_false = ratings_dates.join(
    movies,
    rsuffix='_right',
    how='left'
)
display(joined_false.head())

# исправим: объединим по ключ. столбцу 'movieId' указав его дополнительно в кач-ве индекса для MOVIES (.SET_INDEX)
joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined.head())
```

### метод MERGE ()

- right — присоединяемая таблица. По умолчанию она является «правой».
- how — параметр типа объединения. По умолчанию принимает значение 'inner'.
- on — параметр, который определяет, по какому столбцу происходит объединение. Определяется автоматически, но рекомендуется указывать вручную.
- left_on — если названия столбцов в «левой» и «правой» таблицах не совпадают, то данный параметр отвечает за наименования ключевого столбца исходной таблицы.
- right_on — аналогично предыдущему, параметр отвечает за наименование ключевого столбца присоединяемой таблицы.

```python
# получим ту же таблицу
merged = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
display(merged.head())

# MERGE можно с параметром how='OUTER' использовать вместо CONCAT - дубликаты удаляются автоматически
merge_ratings = rat_1.merge(rat_2, how='outer')
print('Число строк в таблице merge_ratings: ', merge_ratings.shape[0])
display(merge_ratings)
```

```python
""" Напишем ф-ию для извлечени года выпуска фильма """

import re    # библиотека для регулярных выражений
def get_year_release(arg):
    candidates = re.findall(r'\(\d{4}\)', arg)    # находим все слова по шаблону "(DDDD)"
    if len(candidates) > 0:    # проверяем число вхождений
        year = candidates[0].replace('(', '')    # если число вхождений больше 0, очищаем строку от знаков "(" и ")"
        year = year.replace(')', '')
        return int(year)
    else: return None
```

```python
# создадим новый признак - год релиза
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)

# найдём фильм 1999 года с самой низкой средней оценкой
ratings_movies[ratings_movies['year_release'] == 1999].groupby('title')['rating'].mean().sort_values()

# найдём самый непопулярный жанр в 2010 году
ratings_movies[ratings_movies['year_release'] == 2010].groupby('genres')['rating'].mean().sort_values()

# найдём самого всеядного пользователя
ratings_movies.groupby('userId')['genres'].nunique().sort_values(ascending=False)

# найдём пользователя с наименьшим количеством оценок, но наибольшей средней оценкой
ratings_movies.groupby(['userId'])['rating'].agg(
    ['count', 'mean']
    ).sort_values(by=['count', 'mean'], ascending=[True, False])

# фильмы 2018 года с наибольшим рейтингом и минимум 10 оценками
mask = ratings_movies['year_release'] == 2018
grouped_data = ratings_movies[mask].groupby('genres')['rating'].agg(['mean', 'count'])
grouped_data[grouped_data['count'] > 10].sort_values(by='mean', ascending=False)

# создадим новый признак - год выставления оценки
#ratings_movies['date'] = pd.to_datetime(ratings_movies['date'])
ratings_movies['year_rating'] = ratings_movies['date'].dt.year

pivot = ratings_movies.pivot_table(
    values = 'rating',
    index = 'year_rating',
    columns = 'genres',
    aggfunc='mean'
)
display(pivot)

pivot['Action|Adventure'].min()
pivot['Action|Adventure|Animation|Children|Comedy|IMAX'].sort_values(ascending=False)
pivot.loc[2018, :].sort_values(ascending=False).loc['Animation|Children|Mystery']
pivot.Comedy
```
