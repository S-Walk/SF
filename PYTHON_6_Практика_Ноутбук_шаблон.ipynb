**Ваша задача** — вывести минимум и максимум последовательности. Гарантируется, что пользователем будет введено хотя бы одно число.

Для нахождения минимума и максимума напишите функцию `find_min_max`, которая будет возвращать два числа.

**Пример:**


**Подсказка №1:**

Для проверки того, состоит ли строка `s` только из букв, можно воспользоваться методом `s.isalpha()`.

**Подсказка №2:**

Для поиска минимума и максимума в списке можно воспользоваться встроенными функциями `min()` и `max()`.

**Критерии оценивания:**

- Задача решена корректно, учтены все условия (**2 балла**):
    - программа корректно обрабатывает ввод пользователя и работает на любых входных данных;
    - программа корректно обрабатывает присутствие слов во вводе пользователя;

    - решение сопровождено комментариями.
- Задача решена корректно, но не выполнено одно из условий (**1 балл**):
    - программа некорректно обрабатывает ввод пользователя и работает только на конкретном примере входных данных;
    - программа некорректно обрабатывает присутствие слов во вводе пользователя, что приводит к ошибкам при выполнении программы;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): программа некорректно рассчитывает минимум и максимум числовой последовательности.

````python
input_string = input('Введите последовательность чисел: ')

def find_min_max(seq):
    seq = seq.replace(',', '.')    # заменяем запятые на точки
    list_seq = seq.split(' ')    # преобразуем строку в список, разделяя по пробелу
    for elem in list_seq:    # убираем из списка слова
        if elem.isalpha(): list_seq.remove(elem)
    fine_seq = list(map(lambda x: float(x), list_seq))    # перобразуем элементы списка в числа

    return min(fine_seq), max(fine_seq)    # возвращаем минимальное и максимальное значение

#  выводим результат
print(" Minimum: ", find_min_max(input_string)[0], '\n', "Maximum: ", find_min_max(input_string)[1])
````

Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

#### Разложить факториал на простые множители

**Компания:** «СБЕР»

**Уровень:** Сложный

___

Пользователь с клавиатуры (через функцию `input()`) вводит некоторое целое число `n`.

Напишите функцию `decompose_factorial()`, которая будет принимать на вход введённое пользователем число `n`, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.

---

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

Например, факториал числа $3$ будет равен: $3! = 1 * 2 * 3 = 6$.

---

**Простое число** — число, которое делится (нацело) только на себя и на 1. Например, $13$ — простое число, так как оно делится только на себя (на $13$) и на $1$, а $14$ не является простым числом, так как оно делится на $7$ и на $2$.

Подробнее о простых числах можно почитать [здесь](https://dzen.ru/a/XqFMe_VWclT8NTss).

---
**Простые множители** — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа $10$ простыми множителями являются числа $5$ и $2$. Для числа $28$ множителями являются числа $7$ и $4$, но простыми множителями будут числа $7$ и $2$, так как $4$ — это $2 * 2$, а число $28$ представляется в виде $28 = 7 * 2^2$.

----

В результате своей работы функция должна возвращать строку вида (приведён пример для `n=12`):

`12! = 2^10 * 3^5 * 5^2 * 7 * 11`

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

**Пример №1:**
```

5

Вывод:
5! = 2^3 * 3 * 5
```

**Пример №2:**

```
25
```

````python
n = input('Введите число: ')

def decompos_factorial(n):
    n = int(n)
    def fact(n):    # ф-я для поиска факториала числа
        if n == 1 or n == 0: return 1
        res = n * fact(n-1)
        return res
    
    number = fact(n)    # вычисляем факториал
    div_list = []    # создаём пустой список для делителей
    div = 2    # начинаем перебор делителей с 2
    while div * div <= number:    # ограничиваем цикл квадратным корнем от n
        if number % div == 0:    # если делитель подходит, заносим его в список, уменьшаем n в div раз
            div_list.append(div)
            number //= div
        else: div += 1    # если делитель не подходит, берём следующий
    if n > 1: div_list.append(number)    # если в итоге n не превратилось в единицу, и делителей больше нет, значит n - простое число, заносим его в список
    
    div_set = sorted(list(set(div_list)))    # делаем список из уникальных делителей
    res = ''
    for i in div_set:    # ищем кол-во повторений уникальных делителей в исходном списке (степень)
        degree = div_list.count(i)
        if degree == 1: res += str(i) + ' * '    # формируем строку с ответом
        else: res += str(i) + '^' + str(degree) + ' * '
    res = res[:-3]
    return res

print(decompos_factorial(n))
````

## Задание №6 (бонусная, не оценивается)

#### Заменить дублирующиеся буквы

**Компания:** «СБЕР»

**Уровень:** Сложный

____

Пользователь с клавиатуры (через функцию `input()`) вводит строку, состоящую из символов латинского алфавита.

Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.

**Несколько важных моментов:**

* буквы не обязаны быть смежными (смежные — стоящие рядом; в строке `'abc'` буквы a и b, а также b и c являются смежными, в то время как a и c — нет);
* zz заменяем на a;
* вводимая строка состоит из букв латинского алфавита в нижнем регистре.

Напишите функцию `replace_duplicates()`, которая принимает на вход произвольную строку `s`, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

**Пример:**
```
input:
zzzab

step 1: zaba
step 2: zbb
step 3: zc

output:
cz
```

```
input:
hhakafh

step 1: haakfi
step 2: hkfib

output:
hkfib
```

**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string
alphabet = string.ascii_lowercase
print(alphabet)
````

````python
s = input('Введите строку: ')

def replace_duplicates(s, details=False):    # переменная details отвечает за вывод подробной информации
    let_list = list(s)    # преобразуем строку в список
    step = 0    # номер итерации (если details=True)
    while sorted(let_list) != sorted(list(set(let_list))):    # будем запускать цикл по елементам списка пока есть неуникальные элементы
        for let in let_list:    # запускаем цикл по списку букв
            while let_list.count(let) > 1:    # пока есть дубликаты
                if let == 'z': next_let = 'a'    # буква для замены (случай замены z -> a обрабатываем отдельно)
                else: next_let = alphabet[alphabet.index(let) + 1]    
                let_list.remove(let)    # изменяем исходный список чтобы подать его на вход цикла (WHILE - если есть ещё дубликаты текущего елемента FOR - если нужно переходить к следующему элементу)
                let_list.remove(let)
                let_list.append(next_let)
                if details:    # если нужен подробный вывод
                    step += 1
                    print(f'Step {step}: ', ''.join(let_list))
    res = ''.join(let_list)    # формируем результат работы функции
    
    return res

print(replace_duplicates(s, True))
````
