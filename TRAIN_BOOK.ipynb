```python
def swap_places(lst):
    res = []
    res.append(lst[-1])
    res.extend(lst[1:-1])
    res.append(lst[0])
    
    return res

print(swap_places([1, 2, 3]))
## [3, 2, 1]
print(swap_places([1, 2, 3, 4, 5]))
## [5, 2, 3, 4, 1]
print(swap_places(['н', 'л', 'о', 'с']))
## ['с', 'л', 'о', 'н']
```

```python
def equalize_lengths(lst):
    lst.sort(key=len)
    max_string = len(lst[-1])
    lst = lst[::-1]
    res = []
    for string in lst:
        while len(string) < max_string:
            string += '_'
        res.append(string)
    
    return res

print(equalize_lengths(['крот', 'белка', 'выхухоль']))
print(equalize_lengths(['a', 'aa', 'aaa', 'aaaa', 'aaaaa']))
print(equalize_lengths(['qweasdqweas', 'q', 'rteww', 'ewqqqqq']))
```

```python
def find_min_number(a,b,c):
    min_num = a
    if b < min_num: min_num = b
    if c < min_num: min_num = c
    
    return min_num

def sum_min_numbers(a,b,c):
    min_num = find_min_number(a,b,c)
    num_lst = [a,b,c]
    num_lst.remove(min_num)
    if num_lst[0] < num_lst[1]: min_sum = min_num + num_lst[0]
    else: min_sum = min_num + num_lst[1]
    
    return min_sum

print(sum_min_numbers(1, 2, 3))
print(sum_min_numbers(1, 2, -10))
```

```python
matrix1_example = [
          [1, 5, 4, 5],
          [4, 2, -2, -5],
          [4, 2, -2, -5]
]

matrix2_example = [
          [10, 15, 43],
          [41, 2, -2],
          [7, 5, 7]
]

def matrix_sum(matrix1, matrix2):
    if len(matrix1) != len(matrix2):
        print('Error! Matrices dimensions are different!')
        return
    for i in matrix1:
        for j in matrix2:
            if len(i) != len(j):
                print('Error! Matrices dimensions are different!')
                return
    
    res_matr = []
    for i in range(len(matrix1)):
        res_matr.append([])
        for j in range(len(matrix1[i])):
            res_matr[i].append(matrix1[i][j] + matrix2[i][j])
    return res_matr

print(matrix_sum(matrix1=matrix1_example, matrix2=matrix2_example))
```

```python
def get_simple_num(n): 
    lst = []
    need_num = [2,4,6,8]
    for i in range(2, n+1):
        if i > 10:
            if int(str(i)[-1]) in need_num: continue
        for j in lst:
            if j > i ** 1/2:
                lst.append(i)
                break
            if i % j == 0: break
        else: lst.append(i)
    return lst

def fact(n):
    if n == 1 or n == 0: return 1
    res = n * fact(n-1)
    return res

def decompos_factorial(n):
    number = fact(n)
    div_list = []
    div = 2
    while div * div <= number:
        if number % div == 0:
            div_list.append(div)
            number //= div
        else: div += 1
    if n > 1: div_list.append(number)
    
    div_set = sorted(list(set(div_list)))
    res = ''
    for i in div_set:
        degree = div_list.count(i)
        if degree == 1: res += str(i) + ' * '
        else: res += str(i) + '^' + str(degree) + ' * '
    res = res[:-3]
    return res

print(decompos_factorial(22))
```

```python
import string
alphabet = string.ascii_lowercase

input_s = 'eastorwesthomeisbest'

def replace_duplicates(s, details=False):    # переменная details отвечает за вывод подробной информации
    let_list = list(s)    # преобразуем строку в список
    step = 0    # номер итерации (если details=True)
    while sorted(let_list) != sorted(list(set(let_list))):    # будем запускать цикл по елементам списка пока есть неуникальные элементы
        for let in let_list:    # запускаем цикл по списку букв
            while let_list.count(let) > 1:    # пока есть дубликаты
                if let == 'z': next_let = 'a'    # буква для замены (случай замены z -> a обрабатываем отдельно)
                else: next_let = alphabet[alphabet.index(let) + 1]    
                let_list.remove(let)    # изменяем исходный список чтобы подать его на вход цикла (WHILE - если есть ещё дубликаты текущего елемента FOR - если нужно переходить к следующему элементу)
                let_list.remove(let)
                let_list.append(next_let)
                if details:    # если нужен подробный вывод
                    step += 1
                    print(f'Step {step}: ', ''.join(let_list))
    res = ''.join(let_list)    # формируем результат работы функции
    
    return res

print(replace_duplicates(input_s, True))
```
